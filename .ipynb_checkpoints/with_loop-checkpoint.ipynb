```python
# GENERATE COOKIES FILE
def cookies_file_name(email):
    f = email.replace('.', '_').replace('@', '_')
    return f'{f}.txt'


# GET CREDENTIALS
def get_creds(config_file):
    file = open(config_file)
    lines = file.readlines()
    username = lines[0].strip()
    password = lines[1].strip()
    return [username, password]


# INITIAL COOKIE LOADING
def ln_login(driver, config_file):
    driver.get("https://www.linkedin.com/login")
    cred = get_creds(config_file)
    element_id = driver.find_element_by_id("username")
    element_id.send_keys(cred[0])
    element_id = driver.find_element_by_id("password")
    element_id.send_keys(cred[1])
    element_id.submit()
    pickle.dump(driver.get_cookies(), open(cookies_file_name(cred[0]), "wb"))


# LOAD COOKIES
def load_cookies(driver, config_file, url=None):
    cookies_file = cookies_file_name(get_creds(config_file)[0])
    cookies = pickle.load(open(cookies_file, "rb"))
    driver.delete_all_cookies()
    url = "https://www.google.com" if url is None else url
    driver.get(url)
    for cookie in cookies:
        driver.add_cookie(cookie)


def ln_initialize(driver, config_file):
    # DEPENDENCIES: load_cookies, ln_login
    try:
        print("Attempting to load cookies...")
        load_cookies(driver, config_file)
        sleep(1)
        driver.get("https://www.linkedin.com")
        print("Cookies Loaded and entered into your previously logged in account")
        sleep(3)
    except:
        print("Failed to load Cookies. Trying to log in...")
        ln_login(driver, config_file)
        print("Login done.")

        
# SCROLL THE PAGE TO LOAD THE DATA PROPERLY
def scroll_to_load_properly(driver, scroll_pause_time):
    # GET SCROLL HEIGHT
    last_height = driver.execute_script(
        "return document.body.scrollHeight")
    for i in range(5):
        driver.execute_script(
            "window.scrollTo(0, document.body.scrollHeight);")
        sleep(scroll_pause_time)
        driver.execute_script(
            "window.scrollTo(document.body.scrollHeight, document.body.scrollHeight*2/3);")
        sleep(scroll_pause_time)
        driver.execute_script(
            "window.scrollTo(document.body.scrollHeight*2/3, document.body.scrollHeight);")
        sleep(scroll_pause_time)
        # CALCULATE NEW HEIGHT AND COMPARE WITH LAST SCROLL HEIGHT
        new_height = driver.execute_script(
            "return document.body.scrollHeight")
        if new_height == last_height:
            break
        else:
            last_height = new_height
            
            
# click and wait
# def load_and_click(driver, xpath):
#     wait = WebDriverWait(driver, 6)
#     while True:
#         try:
#             el = wait.until(EC.element_to_be_clickable((By.XPATH, xpath)))
#             el.click()
#         except TimeoutException:
#             break  # cannot click the button anymore


# Go To Next preview or Submit
def goto_next(driver):
    try:
        driver.find_element_by_xpath('//button[@aria-label="Review your application"]').click()
    except:
        driver.find_element_by_xpath('//button[@aria-label="Continue to next step"]').click()



# Submit Application
def submit_and_dissmiss(driver):
#     try:
#         try:
    driver.find_element_by_xpath('//button[@aria-label="Submit application"]').click()
#         except:
#             driver.find_element_by_xpath('//button[@data-control-name="submit_application"]').click()
    sleep(2)
    driver.find_element_by_xpath('//button[@aria-label="Dismiss"]').click()
    sleep(1)
#     except:
#         print('Submission failed!')
    
    
    
# Error check
def handle_error(driver):
    while True:
        has_error = False
        try:
            error_els = driver.find_elements_by_class_name('fb-form-element__error-text')
            for error_el in error_els:
                if error_el.get_attribute('class') == 'fb-form-element__error-text t-12':
                    has_error = True
                else:
                    pass
        except:
            pass
        if has_error:
            input('Need your assistance to Fill up the data. Press ENTER whenever you\'re done\n')
        else:
            goto_next(driver)
            break
            
            
            
# Error check 2
def handle_error2(driver, form_name):
    has_error = False
    while True:
        try:
            goto_next(driver)
            easy_apply_content = driver.find_element_by_class_name('jobs-easy-apply-content')
            label = easy_apply_content.find_element_by_tag_name('h3').text.lower()
            if form_name in label:
                has_error = True
            else:
                break
        except:
            error_els = driver.find_elements_by_class_name('fb-form-element__error-text')
            for error_el in error_els:
                if error_el.get_attribute('class') == 'fb-form-element__error-text t-12':
                    has_error = True
                else:
                    pass
        if has_error:
            input('Need your assistance to Fill up the data. Press ENTER whenever you\'re done\n')
        else:
            pass


# Enter Text
def enter_text(element, text):
    element.clear()
    element.send_keys(text)
    
    
# Radio Select
def radio_select(element, label_text):
    label = element.find_element_by_tag_name('label')
    if label_text in label.text.lower():
        label.click()
    else:
        pass
    

def scroll_through(driver):
    for i in range(2):
        try:
            for j in range(1, 26):
    #                 if 25 % j == 0:
    #                     sleep(1)
    #                 else:
    #                     pass
                j_items = driver.find_elements_by_class_name('job-card-list__title')
                element = j_items[j]
                driver.execute_script("arguments[0].scrollIntoView();", element)
        except:
            element = driver.find_element_by_xpath('//section[@aria-label="pagination"]')
            driver.execute_script("arguments[0].scrollIntoView();", element)
```

```python
# in-job functions


# Contact info fill-up
def contact_info(driver):
    try:
        easy_apply_form_data = driver.find_elements_by_class_name('jobs-easy-apply-form-section__grouping')
        for frm in easy_apply_form_data:
            label = frm.find_element_by_tag_name('label').text.strip().lower()
            txt_field = frm.find_element_by_class_name('fb-single-line-text__input')
            if 'first name' in label:
                enter_text(txt_field, first_name)
            elif 'last name' in label:
                enter_text(txt_field, last_name)
            elif 'phone' in label:
                enter_text(txt_field, phone_number)
            else:
                pass
    except:
        pass
        
        
# Home Address
def home_address(element):
    try:
        groups = element.find_elements_by_class_name('jobs-easy-apply-form-section__grouping')
        for group in groups:
            lb = group.find_element_by_tag_name('label').text.lower()
            strt = group.find_element_by_tag_name('input')
            if 'street' in lb:
                enter_text(strt, street)
            elif 'city' in lb:
                enter_text(strt, city)
                sleep(1)
                strt.send_keys(Keys.DOWN)
                strt.send_keys(Keys.RETURN)
            elif 'zip' in lb:
                enter_text(strt, zip_code)
            elif 'state' in lb:
                enter_text(strt, state)
            else:
                pass
    except:
        pass
    
    
# Privacy Policy
def privacy_policy(pb):
    groups = pb.find_elements_by_class_name('jobs-easy-apply-form-section__grouping')
    for p in groups:
        try:
            p.find_element_by_tag_name('label').click()
        except:
            pass
        
        
# Additional Questions
def additional_questions(driver):
    frm_el = driver.find_elements_by_class_name('jobs-easy-apply-form-section__grouping')
    for el in frm_el:
        # Radio check
        try:
            radios = el.find_element_by_class_name('jobs-easy-apply-form-section__element').find_elements_by_class_name('fb-radio')
            for radio in radios:
                label = radio.find_element_by_tag_name('label')
                radio_select(radio, 'yes')
        except:
            pass
        # Questions check
        try:
            question = el.find_element_by_class_name('jobs-easy-apply-form-section__element')
            if 'experience' in question.find_element_by_class_name('fb-form-element-label').text.lower():
                txt_field = question.find_element_by_class_name('fb-single-line-text__input')
                enter_text(txt_field, no_of_years)
                handle_error2(driver, 'experience')
            else:
                pass
        except:
            pass
        
        
# Additional
def additional(pb):
    try:
        groups = pb.find_elements_by_class_name('jobs-easy-apply-form-section__grouping')
        for p in groups:
            # Check radio
            try:
                radios = p.find_elements_by_class_name('fb-radio')
                for radio in radios:
                    label = radio.find_element_by_tag_name('label')
                    enter_text(label, 'yes')
            except:
                pass
    except:
        pass
    
    
# Work authorization
def work_authorization(pb):
    frm_el = pb.find_elements_by_class_name('jobs-easy-apply-form-section__grouping')
    for el in frm_el:
        radios = el.find_element_by_class_name('jobs-easy-apply-form-section__element').find_elements_by_class_name('fb-radio')
        for radio in radios:
            radio_select(radio, 'yes')
            
            
```

```python
def fill_up(driver):
    try:
        easy_apply_content = driver.find_element_by_class_name('jobs-easy-apply-content')
        b4 = easy_apply_content.find_element_by_class_name('pb4')
        pb4 = easy_apply_content.find_elements_by_class_name('pb4')
        for pb in pb4:
            try:
                label = pb.find_element_by_tag_name('h3').text.lower()
                if 'contact info' in label:
                    ################
                    contact_info(driver)
#                     sleep(1)
                    handle_error2(driver, 'contact info')
                    ################
                elif 'home address' in label:
                    ################
                    home_address(pb)
#                     sleep(1)
                    handle_error2(driver, 'home address')
                    ################
                elif 'phone' in label:
                    phone_el = el.find_element_by_class_name('fb-single-line-text__input')
                    enter_text(phone_el, phone_number)
                    handle_error2(driver, 'phone')
                elif 'resume' in label:
                    ################
                    handle_error2(driver, 'resume')
                    ################
                elif 'cover letter' in label:
                    ################
                    handle_error2(driver, 'cover letter')
                    ################
                elif 'voluntary self identification' in label:
                    try:
                        groupings = driver.find_elements_by_class_name('jobs-easy-apply-form-section__grouping')
                        for group in groupings:
                            try:
                                date_picker = group.find_element_by_class_name('artdeco-datepicker__input ')
                                date_picker.clear()
                                date_picker.send_keys(date.today().strftime("%m/%d/%y"))
                            except:
                                pass
                    except:
                        pass
                    input('Need your assistance to fillup the data\nPlease press ENTER in console whenever you\'re done')
                    handle_error2(driver, 'voluntary self identification')
                elif 'photo' in label:
                    #################
                    handle_error2(driver, 'photo')
                    #################
                elif 'work experience' in label:
                    #################
                    driver.find_element_by_xpath('//footer[@role="presentation"]').find_element_by_xpath('//button[@aria-label="Continue to next step"]').click()
                    goto_next(driver)
                    #################
                elif 'education' in label:
                    #################
#                     handle_error2(driver, 'education')
                    driver.find_element_by_xpath('//footer[@role="presentation"]').find_element_by_xpath('//button[@aria-label="Continue to next step"]').click()
                    #################
                elif 'screening questions' in label:
                    handle_error2(driver, 'screening questions')
                elif 'privacy policy' in label:
                    #################
                    privacy_policy(pb)
                    handle_error2(driver, 'privacy policy')
                    #################
                elif 'additional questions' in label:
                    #################
                    additional_questions(driver)
                    handle_error2(driver, 'additional questions')
                    #################
                elif 'additional' in label:
                    ################
                    additional(pb)
                    handle_error2(driver, 'additional')
                    ################
                elif 'work authorization' in label:
                    ################
                    work_authorization(pb)
                    handle_error2(driver, 'work authorization')
                    ################
                else:
                    handle_error2(driver, 'random')
            except:
                pass
    except:
        pass


    
try:
    btn_check = driver.find_element_by_xpath('//footer[@role="presentation"]/div[@class="display-flex justify-flex-end ph5 pv4"]').text.lower()
    if 'back' in btn_check and 'submit application' in btn_check:
        driver.find_element_by_xpath('//button[@aria-label="Back to previous step"]').click()
        goto_next(driver)
        submit_and_dissmiss(driver)
    elif 'back' not in btn_check and 'submit application' in btn_check:
        submit_and_dissmiss(driver)
    else:
        fill_up(driver)
except:
    btn_check2 = driver.find_elements_by_class_name('jobs-apply-form__footer-container')
    if len(btn_check2) > 0 and 'cancel' in btn_check2[0].text.lower() and 'submit application' in btn_check2[0].text.lower():
        try:
            phn = driver.find_element_by_xpath('//input[@class="jobs-apply-form__phone-input "]')
            enter_text(phn, phone_number)
        except:
            pass
        btn_check2[0].find_elements_by_tag_name('button')[1].click()
        driver.find_element_by_xpath('//button[@aria-label="Dismiss"]').click()
        sleep(1)
    else:
        fill_up(driver)
```
